```
import matplotlib.pyplot as plt
import math

def g1(xn, criterio_paro = 0.0001):
    No_Conv=0
    converge = True
    mensaje = 'Converge'
    x = [xn]
    Error=[]
    n=0
    x.append((2*math.pow(x[n], 2))-5)
    
    n+=1
    Error.append(abs(x[n] - x[n-1]))
    
    while ((Error[n-1] > criterio_paro) and No_Conv<50 and x[n]<999999999999):
            
            x.append((2*math.pow(x[n], 2))-5)
            n+=1
            No_Conv+=1
            Error.append(abs(x[n] - x[n-1]))
            if(No_Conv==50 or x[n]>999999999999):
                mensaje = 'No converge'
                converge = False
            
            
    print('g1')
    print('x: '+ str(x) )
    print('\n Error: ' + str(Error))
   
    plt.plot(x,color='b')
    plt.plot(Error,color='r')
    plt.title("g1 " + mensaje, fontsize = 24)
    plt.xlabel("Rojo = Vector del Error", fontsize = 14)
    plt.ylabel("Azul = Raiz", fontsize = 14)
    plt.show()
    
    return converge

def g2(xn, criterio_paro = 0.0001):
    
    No_Conv=0
    converge = True
    mensaje = 'Converge'
    x = [xn]
    Error=[]
    n=0
    x.append(5/(2*x[n]-1))
    n+=1
    Error.append(abs(x[n] - x[n-1]))
    
    while (Error[n-1] > criterio_paro and No_Conv<50 and x[n]<999999999999):
            
            x.append(5/(2*x[n]-1))
            n+=1
            No_Conv+=1
            Error.append(abs(x[n] - x[n-1]))
            if(No_Conv==50 or x[n]>999999999999):
                mensaje = 'No converge'
                converge = False
            
            
    print('\n \n g2')
    print('x: '+ str(x) )
    print('\n Error: ' + str(Error))
    plt.plot(x,color='b')
    plt.plot(Error,color='r')
    plt.title("g2 " + mensaje, fontsize = 24)
    plt.xlabel("Rojo = Vector del Error", fontsize = 14)
    plt.ylabel("Azul = Raiz", fontsize = 14)
    plt.show()
    
    return converge
    
    

def g3(xn, criterio_paro = 0.0001):
    
    No_Conv=0
    converge1 = True
    mensaje1 = 'Converge'
    #Positivo
    x1 = [xn]
    Error1=[]
    n=0
    x1.append(math.sqrt((x1[n]+5)/2))
    n+=1
    Error1.append(abs(x1[n] - x1[n-1]))
    
    while ((Error1[n-1] > criterio_paro) and No_Conv<50 and x1[n]<999999999999):
            
            x1.append(math.sqrt((x1[n]+5)/2))
            n+=1
            No_Conv+=1
            Error1.append(abs(x1[n] - x1[n-1]))
            if(No_Conv==50 or x1[n]>999999999999):
                mensaje1 = 'No converge'
                converge1 = False
            
    print('\n \n g3 Positivo')        
    print('x: '+ str(x1) )
    print('\n Error: ' + str(Error1))
    plt.plot(x1,color='b')
    plt.plot(Error1,color='r')
    plt.title("g3 positivo " + mensaje1, fontsize = 24)
    plt.xlabel("Rojo = Vector del Error", fontsize = 14)
    plt.ylabel("Azul = Raiz", fontsize = 14)
    plt.show()
     
    
    No_Conv=0
    converge2 = True
    mensaje2 = 'Converge'
    #Negativo
    x2 = [xn]
    Error2=[]
    n=0
    x2.append((math.sqrt((x2[n]+5)/2))*-1)
    n+=1
    Error2.append(abs(x2[n] - x2[n-1]))
    
    
    while ((Error2[n-1] > criterio_paro) and No_Conv<50 and x2[n]<999999999999):
            
            x2.append((math.sqrt((x2[n]+5)/2))*-1)
            n+=1
            No_Conv+=1
            Error2.append(abs(x2[n] - x2[n-1]))
            if(No_Conv==50 or x2[n]>999999999999):
                mensaje2 = 'No converge'
                converge2 = False
                
    print('\n \n g3 Negativo')   
    print('x: '+ str(x2) )
    print('\n Error: ' + str(Error2))
    plt.plot(x2,color='b')
    plt.plot(Error2,color='r')
    plt.title("g3 negativo " + mensaje2, fontsize = 24)
    plt.xlabel("Rojo = Vector del Error", fontsize = 14)
    plt.ylabel("Azul = Raiz", fontsize = 14)
    plt.show()
    
    return converge1 + converge2
    


xn = int(input('Introduce el valor incial de x \n'))
s=int(input("Selecciona 1 si quieres introducir un criterio de paro de lo contrario escribe cualquier numero \n"))
if s==1:
    criterio_paro = float(input('Introduce el valor del criterio de paro \n'))
    print(str(g1(xn,criterio_paro)))
    print(str(g2(xn,criterio_paro)))
    print(str(g3(xn,criterio_paro)))
else:    
    print(str(g1(xn)))
    print(str(g2(xn)))
    print(str(g3(xn)))
```
